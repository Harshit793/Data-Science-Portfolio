```python
def find_number(text):
    num =re.findall(r'[0-9]+',text)
    return " ".join(num)
```

```python
#https://towardsdatascience.com/stacking-classifiers-for-higher-predictive-performance-566f963e4840
from mlxtend.classifier import StackingCVClassifier 
sclf=StackingCVClassifier(classifiers=[clf3,clf4,clf5,clf6],cv=5,shuffle=False,meta_classifier=clf2) 
```

```python
classifiers1={"SVC":clf3,"DT":clf4,"RF":clf5,"XGB":clf6,"Stack":sclf}
```

```python
# Train classifiers
for key in classifiers1:
    # Get classifier
    classifier=classifiers1[key]
    
    # Fit classifier
    classifier.fit(X_tr_encoded.values,Y_tr.values)
        
    # Save fitted classifier
    classifiers1[key]=classifier
```

```python
# Get results
results = pd.DataFrame()
for key in classifiers1:
    # Make prediction on test set
    if key=='SVC':
      y_pred=classifiers1['SVC'].decision_function(X_val_encoded.values)
    # Save results in pandas dataframe object
      results[f"{key}"] = y_pred
    else: 
      y_pred=classifiers1[key].predict_proba(X_val_encoded.values)[:,1]
      results[f"{key}"] = y_pred
```

```python
_# Add the test set to the results object
results["Target"]=Y_val.reset_index(drop=True)
results
```

```python
X=results.drop('Target',axis=1)
Y=results['Target']
for i in X.columns:
  score=roc_auc_score(Y_val,X[i])
  print("Auc score of {} is".format(i),score)
```
